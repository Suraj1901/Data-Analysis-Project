```python
import pandas as pd

file = r'C:\Users\suraj\Downloads\Suraj Documents\Project\Oasis Infobyte\Projects Datasets\Cleaning Data\AB_NYC_2019 Dataset 1.csv'

df = pd.read_csv(file)
df.info()
```

```python
# Drop rows with missing 'name' or 'host_name' since they're critical

df = df.dropna(subset=['name', 'host_name'])
```

```python
df.isnull().sum()
```

```python
# Fill missing 'reviews_per_month' with 0 (assuming no reviews)

df['reviews_per_month'].fillna(0)
```

```python
# Fill missing 'last_review' with a placeholder date (or could be left as NaT if needed)

df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
df
```

```python
df.info()
```

```python
#Remove duplicates
df = df.drop_duplicates()
```

```python
#Standardize column names (lowercase and underscores)
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df
```

```python
#Outlier detection and handling
# Define thresholds for outliers (domain knowledge based)df= df[
    (df['price'] <= 1000) &
    (df['minimum_nights'] <= 365) &
    (df['availability_365'] <= 365)
]

df
```
